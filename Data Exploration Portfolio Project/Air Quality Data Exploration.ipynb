# **2. Data Cleaning**

Check missing or junk data in our dataframe

```python
df.isna()
```

Remove columns with NaN values

```python
clean_df = df.dropna(axis=1)
```

```python
clean_df.head()
```

Get a concise summary of the DataFrame

```python
print("DataFrame Info:")
clean_df.info()
print("\n" + "="*50 + "\n")
```

Get descriptive statistics for numerical columns

```python
print("Descriptive Statistics:")
print(clean_df.describe())
print("\n" + "="*50 + "\n")
```

# **3. Exploring Key Categorical Columns**

Explore 'Name' (the specific gas being measured)

```python
print("Unique Gases ('Name'):")
print(clean_df['Name'].unique())
print("\nValue counts for 'Gas Name':")
print(clean_df['Name'].value_counts())
print("\n" + "="*50 + "\n")
```

Explore 'Measure' (the unit of measurement)

```python
print("Unique Measures:")
print(clean_df['Measure'].unique())
print("\nValue counts for 'Measure':")
print(clean_df['Measure'].value_counts())
print("\n" + "="*50 + "\n")
```

Explore 'Geo Type Name' (e.g., Borough, Citywide)

```python
print("Unique Geo Type Names:")
print(clean_df['Geo Type Name'].unique())
print("\nValue counts for 'Geo Type Name':")
print(clean_df['Geo Type Name'].value_counts())
print("\n" + "="*50 + "\n")
```

Explore 'Time Period'

```python
print("Unique Time Periods:")
print(clean_df['Time Period'].unique())
print("\nValue counts for 'Time Period':")
print(clean_df['Time Period'].value_counts())
print("\n" + "="*50 + "\n")
```

# **4. Exploring the Date Column ('Start_Date')**

```python
# Convert 'Start_Date' to datetime objects
clean_df['Start_Date'] = pandas.to_datetime(clean_df['Start_Date'])

# Explore the date range
print("Earliest Start Date:", clean_df['Start_Date'].min())
print("Latest Start Date:", clean_df['Start_Date'].max())
print("\n" + "="*50 + "\n")
# Extract year and month for further analysis
clean_df['Year'] = clean_df['Start_Date'].dt.year
clean_df['Month'] = clean_df['Start_Date'].dt.month

print("Value counts for 'Year':")
print(clean_df['Year'].value_counts())
print("\nValue counts for 'Month':")
print(clean_df['Month'].value_counts())
```

# **Examining the Numerical Data Value:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of 'Data Value'
plt.figure(figsize=(10, 6))
sns.histplot(clean_df['Data Value'], kde=True)
plt.title('Distribution of Data Value')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.show()

```

```python
# Box plot of 'Data Value' to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=clean_df['Data Value'])
plt.title('Box Plot of Data Value')
plt.ylabel('Data Value')
plt.show()
```

**Exploring Relationships (Initial Look) - Focusing on gas name 'Name'**

```python
# 'Data Value' by 'Geo Type Name'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Geo Type Name', y='Data Value', data=clean_df)
plt.title('Data Value by Geo Type Name')
plt.xlabel('Geo Type Name')
plt.ylabel('Data Value')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate the mean 'Data Value' for each 'Name'
mean_data_by_gas = clean_df.groupby('Name')['Data Value'].mean().reset_index()
# Create the bar chart
plt.figure(figsize=(30,15))
sns.barplot(x='Name', y='Data Value', data=mean_data_by_gas)
plt.title('Average Data Value by Gas Name')
plt.xticks(rotation='vertical')
plt.xlabel('Name', rotation='vertical')
plt.ylabel('Average Data Value')
plt.tight_layout()
plt.show()
```

```python
# Explore 'Data Value' by 'Name' and 'Geo Type Name'
plt.figure(figsize=(15, 8))
plt.xticks(rotation='vertical')
sns.boxplot(x='Name', y='Data Value', hue='Name', data=clean_df)
plt.title('Data Value by Geo Type Name, Colored by (Gas) Name')
plt.xlabel('Geo Type Name')
plt.ylabel('Data Value')
plt.legend(title='Name')
plt.show()
```
